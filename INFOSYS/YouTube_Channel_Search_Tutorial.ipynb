## ğŸ” Step 3: Finding a YouTube Channel

**The Problem:** YouTube uses special IDs (like "UCsT0YIqwnpJCM-mx7-gSA4Q") for channels, not names!

**The Solution:** We need a function that converts a channel name (like "TED") into its ID.

**Real-world example:** It's like finding someone's phone number from their name in a contact list! ğŸ“±

Below is a simple function that does this for us:

````python
def get_channel_id(channel_name):
    """
    ğŸ¯ This function finds a YouTube channel's ID from its name
    
    Example: "TED" â†’ "UCsT0YIqwnpJCM-mx7-gSA4Q"
    """
    try:
        # Ask YouTube: "Hey, where is this channel?"
        request = youtube.search().list(
            part='snippet',           # Get basic info
            q=channel_name,           # The channel name we're looking for
            type='channel',           # We want a channel, not a video
            maxResults=1              # Just give us the best match
        )
        response = request.execute()  # Send the request!
        
        # Did we find it?
        if response['items']:
            channel_id = response['items'][0]['snippet']['channelId']
            print(f"âœ… Found '{channel_name}'!")
            print(f"   ğŸ†” Channel ID: {channel_id}")
            return channel_id
        else:
            print(f"âŒ Sorry, couldn't find '{channel_name}'")
            return None
            
    except Exception as e:
        print(f"âŒ Error: {e}")
        return None

print("âœ… Function 'get_channel_id' is ready to use!")
````

## ğŸ¥ Step 4: Search for Videos Inside a Channel

Now that we know how to find a channel, let's search for specific videos inside it!

**Example:** Search for "artificial intelligence" videos in the TED channel.

**How it works:**
1. We give it a search term (like "AI" or "cooking")
2. We give it the channel ID
3. It returns a list of matching videos!

It's like searching for a specific book in a specific library! ğŸ“š

````python
def search_channel_videos(query, channel_id, max_results=5):
    """
    ğŸ¯ Search for videos inside a specific YouTube channel
    
    Inputs:
    - query: What you want to search for (e.g., "python tutorial")
    - channel_id: The channel's ID (we got this from the previous function)
    - max_results: How many videos to return (default: 5)
    
    Output: A list of videos with their info
    """
    try:
        # Ask YouTube: "Find videos about [query] in channel [channel_id]"
        request = youtube.search().list(
            part='snippet',              # Get video details
            q=query,                     # What we're searching for
            channelId=channel_id,        # ONLY search in this channel!
            type='video',                # We want videos, not playlists
            maxResults=max_results,      # How many results
            relevanceLanguage='en',      # Prefer English results
            videoCaption='any'           # Videos with or without captions
        )
        response = request.execute()
        
        # Extract the video information
        videos = []
        for item in response.get('items', []):
            video_info = {
                'video_id': item['id']['videoId'],
                'title': item['snippet']['title'],
                'channel': item['snippet']['channelTitle'],
                'description': item['snippet']['description']
            }
            videos.append(video_info)
        
        print(f"âœ… Found {len(videos)} videos about '{query}'!")
        return videos
        
    except Exception as e:
        print(f"âŒ Error: {e}")
        return []

print("âœ… Function 'search_channel_videos' is ready!")
````

### ğŸ§ª Let's Try It! Search for AI Videos in TED

Let's search for "artificial intelligence" videos in the TED channel:

````python
# Search for AI videos in TED
print("ğŸ” Searching for 'artificial intelligence' videos in TED channel...")
print()

videos = search_channel_videos("artificial intelligence", ted_channel_id, max_results=5)

# Show the results in a nice format
print("\n" + "="*60)
print("ğŸ“º RESULTS:")
print("="*60)

for i, video in enumerate(videos, 1):
    print(f"\nğŸ¬ Video #{i}")
    print(f"Title: {video['title']}")
    print(f"Watch: https://www.youtube.com/watch?v={video['video_id']}")
    print(f"About: {video['description'][:100]}...")
    print("-" * 60)
````

## ğŸ“ Step 5: Check if Videos Have Subtitles/Transcripts

Sometimes you want to read what's being said in a video (subtitles/transcripts). Let's check which videos have them!

**Why is this useful?**
- âœ… You can read along while watching
- âœ… You can translate to other languages
- âœ… You can search for specific words in the video
- âœ… Great for learning and taking notes!

Let's create a function to check this:

````python
def check_transcript(video_id):
    """
    ğŸ¯ Check if a video has English subtitles/transcripts
    
    Input: video_id (the video's ID)
    Output: 
    - has_transcript: True if subtitles exist, False if not
    - is_auto: True if auto-generated, False if manually added
    - transcript_text: The actual subtitle text (or None)
    """
    try:
        # Get list of all available transcripts for this video
        transcript_list = YouTubeTranscriptApi.list_transcripts(video_id)
        
        # Try to find MANUAL English subtitles first (these are better quality)
        try:
            transcript = transcript_list.find_transcript(['en'])
            transcript_data = transcript.fetch()
            # Combine all subtitle lines into one text
            transcript_text = " ".join([entry['text'] for entry in transcript_data])
            return True, False, transcript_text  # Found manual subtitles!
        except:
            pass  # Didn't find manual, let's try auto-generated
        
        # Try to find AUTO-GENERATED English subtitles
        try:
            transcript = transcript_list.find_generated_transcript(['en'])
            transcript_data = transcript.fetch()
            transcript_text = " ".join([entry['text'] for entry in transcript_data])
            return True, True, transcript_text  # Found auto-generated!
        except:
            pass  # No auto-generated either
        
        # No English subtitles found at all
        return False, False, None
        
    except Exception as e:
        # Something went wrong
        return False, False, None

print("âœ… Function 'check_transcript' is ready!")
````

### ğŸ§ª Check Which Videos Have Subtitles

Let's check all the videos we found to see which ones have English subtitles:

````python
# Check each video for subtitles
print("ğŸ” Checking which videos have English subtitles...")
print()

for i, video in enumerate(videos, 1):
    print(f"\nğŸ“¹ Video #{i}: {video['title'][:50]}...")
    
    # Check for subtitles
    has_transcript, is_auto, transcript_text = check_transcript(video['video_id'])
    
    if has_transcript:
        if is_auto:
            print("   âœ… Has AUTO-GENERATED English subtitles")
        else:
            print("   âœ… Has MANUAL English subtitles (high quality!)")
        
        # Show a preview of the subtitles (first 100 characters)
        print(f"   Preview: {transcript_text[:100]}...")
    else:
        print("   âŒ No English subtitles available")
    
    print("-" * 60)
````

## ğŸš€ Step 6: The Complete System (All-In-One!)

Now let's combine everything into ONE easy-to-use tool!

**What it does:**
1. Takes a channel name (like "TED")
2. Takes what you want to search for (like "climate change")
3. Finds the videos
4. Optionally filters to show ONLY videos with subtitles
5. Returns the results!

This is like having a smart assistant that does all the work for you! ğŸ¤–

````python
class YouTubeChannelSearcher:
    """
    ğŸ¯ Your YouTube Search Assistant!
    
    This is a complete tool that can:
    - Find any YouTube channel
    - Search videos in that channel
    - Check for subtitles
    - Filter results
    """
    
    def __init__(self, api_key):
        """Initialize - connect to YouTube"""
        self.youtube = build('youtube', 'v3', developerKey=api_key)
        print("âœ… YouTube Search Assistant ready!")
    
    def get_channel_id(self, channel_name):
        """Find a channel's ID from its name"""
        try:
            request = self.youtube.search().list(
                part='snippet',
                q=channel_name,
                type='channel',
                maxResults=1
            )
            response = request.execute()
            
            if response['items']:
                return response['items'][0]['snippet']['channelId']
            return None
        except:
            return None
    
    def search_videos(self, query, channel_id, max_results=5, only_with_transcripts=False):
        """
        Search for videos in a channel
        
        Inputs:
        - query: What to search for
        - channel_id: Which channel to search in
        - max_results: How many videos (default: 5)
        - only_with_transcripts: True = only show videos with subtitles
        """
        try:
            # If filtering by transcripts, get more videos to check
            search_limit = max_results * 4 if only_with_transcripts else max_results
            
            request = self.youtube.search().list(
                part='snippet',
                q=query,
                channelId=channel_id,
                type='video',
                maxResults=search_limit,
                relevanceLanguage='en'
            )
            response = request.execute()
            
            # Collect video information
            videos = []
            for item in response.get('items', []):
                video = {
                    'video_id': item['id']['videoId'],
                    'title': item['snippet']['title'],
                    'channel': item['snippet']['channelTitle'],
                    'description': item['snippet']['description']
                }
                videos.append(video)
            
            # Filter by subtitles if requested
            if only_with_transcripts:
                print("ğŸ” Filtering for videos with English subtitles...")
                filtered_videos = []
                for video in videos:
                    has_transcript, is_auto, _ = self.check_transcript(video['video_id'])
                    if has_transcript:
                        video['has_transcript'] = True
                        video['is_auto_generated'] = is_auto
                        filtered_videos.append(video)
                        if len(filtered_videos) >= max_results:
                            break
                return filtered_videos
            
            return videos[:max_results]
            
        except Exception as e:
            print(f"âŒ Error: {e}")
            return []
    
    def check_transcript(self, video_id):
        """Check if video has English subtitles"""
        try:
            transcript_list = YouTubeTranscriptApi.list_transcripts(video_id)
            
            # Try manual first
            try:
                transcript = transcript_list.find_transcript(['en'])
                transcript_data = transcript.fetch()
                text = " ".join([entry['text'] for entry in transcript_data])
                return True, False, text
            except:
                pass
            
            # Try auto-generated
            try:
                transcript = transcript_list.find_generated_transcript(['en'])
                transcript_data = transcript.fetch()
                text = " ".join([entry['text'] for entry in transcript_data])
                return True, True, text
            except:
                pass
            
            return False, False, None
        except:
            return False, False, None
    
    def search_channel(self, channel_name, query, max_results=5, only_with_transcripts=False):
        """
        ğŸ¯ MAIN FUNCTION - Do everything in one go!
        
        Just give it:
        - channel_name: Like "TED" or "Khan Academy"
        - query: Like "python" or "climate change"
        - max_results: How many videos you want
        - only_with_transcripts: True if you only want videos with subtitles
        """
        print(f"\nğŸ” Step 1: Finding '{channel_name}' channel...")
        channel_id = self.get_channel_id(channel_name)
        
        if not channel_id:
            print(f"âŒ Couldn't find channel: {channel_name}")
            return []
        
        print(f"âœ… Found it!")
        print(f"\nğŸ” Step 2: Searching for '{query}' videos...")
        videos = self.search_videos(query, channel_id, max_results, only_with_transcripts)
        print(f"âœ… Found {len(videos)} videos!\n")
        
        return videos

# Create our YouTube search assistant!
searcher = YouTubeChannelSearcher(api_key)
print("ğŸ‰ Your search assistant is ready to use!")
````

## Summary: How the Model Works

### Architecture Overview:

1. **YouTube Data API v3**: Core search functionality
   - `search().list()` - Search for channels and videos
   - `channelId` parameter - Filter videos to specific channel
   - `q` parameter - Search query

2. **Channel ID Resolution**: 
   - User enters channel name â†’ API searches â†’ Returns channel ID
   - Channel ID is required for filtering videos

3. **Video Search**:
   - Search with query + channel ID
   - Returns video metadata (title, description, video_id)
   - Can request more results if filtering needed

4. **Transcript Checking**:
   - `youtube-transcript-api` library
   - Checks for manual English transcripts first
   - Falls back to auto-generated if needed
   - Some videos block API access

5. **Filtering**:
   - If transcript filter enabled: check each video
   - Stop when required number found
   - Return filtered list

### Key Advantages:
- âœ… Search within specific channels only
- âœ… Find videos with English transcripts
- âœ… Real-time search (no pre-downloading)
- âœ… Works with any YouTube channel
