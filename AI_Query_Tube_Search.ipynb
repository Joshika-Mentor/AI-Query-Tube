```python
from googleapiclient.discovery import build
import pandas as pd
import time
```

build() helps you create a connection to YouTube API.
time : This helps us â€œpauseâ€ the code for a short time., Because making too many API calls quickly can cause quota issues.

```python
def get_all_videos(api_key, channel_id):
    youtube = build("youtube", "v3", developerKey=api_key)
    videos = []
    next_page_token = None

    while True:
        req = youtube.search().list(
            part="snippet",
            channelId=channel_id,
            maxResults=50,
            order="date",
            type="video",
            pageToken=next_page_token
        )

        response = req.execute()

        for item in response.get("items", []):
            videos.append({
                "video_id": item["id"]["videoId"],
                "title": item["snippet"]["title"],
                "published_at": item["snippet"]["publishedAt"]
            })

        next_page_token = response.get("nextPageToken")
        if not next_page_token:
            break

        time.sleep(0.1)

    return pd.DataFrame(videos)


# âœ… Call the function to create df
df = get_all_videos(API_KEY, CHANNEL_ID)

# âœ… Show first 5 rows
df.head()
```
